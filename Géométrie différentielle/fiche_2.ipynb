# Exercice 3          
On va représenter l'application $ f: U = \mathbb R \times ]0,+\infty[ ~\longrightarrow ~\mathbb R ^3 $          
définie par $ f (u,v)  \longrightarrow \begin{pmatrix} e^{-v} • \cos u \\ e^{-v} • \sin u \\ v  \end{pmatrix}$

```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python
def f(u,v):
    f_1 = np.exp(-v)*np.cos(u)
    f_2 = np.exp(-v)*np.sin(u)
    f_3 = v
    
    return f_1 , f_2 , f_3 
```

```python
up = np.linspace(0, 5, 100)
vp = np.linspace(0, 2 * np.pi, 100)
up , vp = np.meshgrid(up, vp)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

f_1p , f_2p , f_3p = f(up,vp)

ax.plot_surface(f_1p, f_2p, f_3p , cmap='viridis')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.show()
```
