### Question 4
- Afficher le nuage de points qui montre la relation entre tip et total_bill


```python
df.plot(x='tip',y='total_bill',kind='scatter')
plt.show()
```

### Question 5
- Afficher la relation entre total_bill, tip et size


```python
import mpl_toolkits.mplot3d.axes3d as axes3d

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['total_bill'], df['tip'], df['size'])

ax.set_xlabel('total_bill')
ax.set_ylabel('tip')
ax.set_zlabel('axes3d')

plt.show()
```

### Question 6
- Afficher la relation entre days et total_bill

### Question 7
- Afficher le nuage de points avec les jours en y et tip en x. Mettre une couleur différente en fonction de sex


```python
color_map = {'Female': 'red', 'Male': 'blue'}
df['sex'] = df['sex'].map(color_map)

df.plot(x='tip',y='day',kind='scatter',c=df['sex'])
```
