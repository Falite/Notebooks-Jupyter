```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x,n):
    return 1/(x**n+np.exp(x))

xp=np.linspace(0,10,5000)

def g(x):
    return 1/np.exp(x)
yp=np.linspace(0,1,1000)
```

```python
plt.figure(figsize=(7,15))
for i in range(1,10):
    plt.plot(xp,f(xp,i),'b')
plt.plot(yp,g(yp),'r')
plt.xlabel('x')
plt.ylabel('fn(x)')
plt.title('graphes des fonctions exo2 partiel1 2018-2019')
plt.show()
```
