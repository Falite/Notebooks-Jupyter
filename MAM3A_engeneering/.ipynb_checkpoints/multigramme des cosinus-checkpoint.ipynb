```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, n):
    return (np.cos(np.pi * x)) ** n

def I_pairs(x):
    return np.where(x % 2 == 0, 1, 0)

xp = np.linspace(-np.pi, np.pi, 1000)
plt.figure(figsize=(15, 7))
plt.ylabel("cosinus")
plt.xlabel("abscisses")
for n in range(1, 200):
    plt.plot(xp, f(xp, n), 'b', alpha=0.5) 
plt.plot(xp, I_pairs(xp), 'r', label='Indicatrice des nombres pairs')
plt.legend()
plt.title("Multigraphe des cos(pi*x)^n et indicatrice des nombres pairs")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Créez une figure
plt.figure(figsize=(20,10))
fig, ax = plt.subplots()

# Créez une ellipse en spécifiant les coordonnées du centre (x, y), la largeur, la hauteur et l'angle d'inclinaison
ellipse = Ellipse((0, 0), width=5, height=3, angle=0)

# Ajoutez l'ellipse à l'axe
ax.add_patch(ellipse)

# Réglez les limites de l'axe pour afficher l'ellipse
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)

# Réglez l'aspect de l'axe pour que les unités sur les deux axes soient égales
ax.set_aspect('equal')

# Affichez la figure

plt.xlabel('Axe X')
plt.ylabel('Axe Y')
plt.title('Représentation d\'une ellipse')
plt.grid(True)
plt.show()

```
