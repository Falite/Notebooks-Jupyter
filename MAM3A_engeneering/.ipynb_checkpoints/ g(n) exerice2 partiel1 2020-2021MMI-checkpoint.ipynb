```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return np.exp(-x) #exemple de fonction int√©grable Lebesgue 


def h(x,n):
    return np.exp(-(x-1)*n)*f(x)


xp=np.linspace(1,3,100)
```

```python
for i in range(1,30):
    plt.plot(xp,h(xp,i),'b',alpha=0.5)
plt.xlabel('x')
plt.ylabel('hn(x)')
plt.title('g(n) exercice2 partiel1 2020-2021')
plt.show()
```
