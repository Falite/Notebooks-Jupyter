we have $ f_n(x)= \frac{\sin(\pi x)}{x^n + 1},~~\forall x \geq 0 $       
           
and we have $f(x)=\binom{\sin(\pi x) ~~~\text{if} ~~~~x \in [0,1]}{0 ~~~\text{if}~~~~x>1}$          
         
we want to show that $ \underset{n \rightarrow + \infty }{\text{lim}} f_n(x)=f(x)$

```python
import numpy as np
import matplotlib.pyplot as plt 
import math
```

```python
def f(x,n):
    return np.sin(np.pi*x)/((x**n)+1)

xp=np.linspace(0,10,500)

def g(x):
    return np.sin(np.pi*x)
yp=np.linspace(0,1,500)
```

```python
for i in range(2,20):
    plt.plot(xp,f(xp,i),'b',alpha=0.5)
plt.plot(yp,g(yp),'r')
plt.title('exercice 2 partiel1 2019-2020')
plt.xlabel('x')
plt.ylabel('fn(x)')
plt.show()
```
