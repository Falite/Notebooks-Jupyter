```python
import numpy as np  
import pandas as pd
```

```python
"""1) Lire les donn´ees `a l’aide de la biblioth`eque python pandas et faire un r´esum´e des donn´ees
(combiens de donn´ees, combien de classes, r´epartition par classe, ”range” des donn´ees, histogramme des valeurs...)"""


iris = pd.read_csv('https://datahub.io/machine-learning/iris/r/iris.csv');
iris
```

```python
iris.info() 
```

```python
iris[0:10] 
```

```python
"""2) Extraire de ce fichier 2 tableaux de type numpy, le premier appeler X avec 4 colonnes et
autant de ligne que de mesures et le second appel´e Y un tableau de label de classe `a prendre
parmi les valeurs 0,1 ou 2 """

X= iris.iloc[:, :-1].values

#iris['class']=iris['class'].astype('category')
#Y=iris['class'].cat.codes.values

Y = iris.iloc[:,-1].values
Y= np.where(Y=='Iris-setosa',0 , np.where(Y=='Iris-versicolor',1,2))

print(X)
print(Y)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

```

```python

model = keras.Sequential()

model.add(layers.Input(shape=(4,)))

model.add(layers.Dense(8, activation='relu'))

model.add(layers.Dense(1, activation='linear'))
```

```python
model.get_weights()
```

```python
# Compiler le modèle en spécifiant la fonction de perte, l'optimiseur
model.compile(loss='mse', optimizer='adam')
```

```python
# Entraîner le modèle sur les données d'entraînement
model.fit(X, Y, epochs=150, batch_size=150, verbose=2)
```

```python
Ypred=model.predict(X)
Ypred.round()
```

```python
Matrice_confusion=tf.math.confusion_matrix(Y,Ypred)
#On affiche la matrice de confusion
```
