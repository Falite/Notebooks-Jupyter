```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

we want to approach $\int_{-1}^1 \exp(-x^2) dx$ using the compound trapezoid method         
            
we note $T_n$ the method of compound trapezoids applied to the integral        
          
we calculate the error knowing $ \Big| \int_{-1}^1 \exp(-x^2) dx - T_n \Big| \leq \frac 43 \times \frac 1{n^2}$       
          
we want to know for which n the error is less than $10^{-4}$

```python
def E(n):
    return (4/3)*((1/n)**2)
```

```python
for i in range(1,200):
    print("l'erreur est majorée par",round(E(i),6),"pour i=",i)
```

```python
xp=np.linspace(1,100)
plt.plot(xp,E(xp),'b')
plt.title("courbe de l'erreur")
plt.ylabel("Erreur")
plt.xlabel("en fonction de n")
plt.show()
```
