```python
import numpy as np 
import matplotlib.pyplot as plt
```

On va repr√©senter $f_n (t) = t^n $ sur $\mathcal F \subset \mathcal C( [0,1],\mathbb R),~~\mathcal F = \{f_n,~n \geq 0\}$            
On appelle $f_n$ "f(t,n)"        
On fixe les bornes du compact "limg" = $0$ et "limd" = $1$

```python
def f(t,n):
    return t**n 

limg=0
limd=1

xp=np.linspace(limg,limd,100)
```

```python
plt.figure(figsize=(10,6))
        
for i in range(0,11):
    plt.plot(xp,f(i,xp), label=f'$f_{i}(t) = t^{i}$')
    
plt.title('Courbes $f_n(t) = t^n$')
plt.xlabel('t')
plt.ylabel('$f_n(t)$')
plt.yticks(range(0,10))
plt.xlim(0,1)
plt.legend()
plt.show
```
