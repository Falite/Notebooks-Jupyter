```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
xp=np.linspace(-5,5,1000)
```

```python
def Phi(x):
    return 1/2*np.cos(x)
```

```python
def Biss(x):
    return x
```

```python
plt.plot(xp,Phi(xp),'b')
plt.plot(xp,Biss(xp),'r')
plt.ylim(-1,1)
plt.show()
```
