```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (1500-20*x)/15

def p(x):
    return (800-10*x)/5

def a(x):
    return (2000-20*x)/25

def m(x):
    return (4500*x)/2500

L=[f,p,a]
```

```python
xp=np.linspace(0,2000,50000)
```

```python
for i in L:
    plt.plot(xp,i(xp),'b')
plt.plot(xp,m(xp),'r')
plt.title('probl√®me exo 5')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
