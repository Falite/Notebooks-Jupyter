```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
xp=np.linspace(-1,1,100)
def y1(t,k):
    return 1/(1-k*t)
```

```python
for k in range(10):
    plt.plot(xp,y1(xp,k),'b',alpha=0.5)
plt.xlabel('variables dans R')
plt.ylabel('solutions')
plt.title('courbes de l\'exo 1')
plt.show()
```

```python
yp=np.linspace(-5,5,100)
def y2(t):
    return np.sqrt((t**2-1)/3)+t**2-1-np.sqrt(3*(t**2-1))
```

```python
plt.plot(yp,y2(yp),'r')
plt.ylabel('courbe de l\'exo 2')
plt.xlabel('valeurs dans R')
plt.show()
```
