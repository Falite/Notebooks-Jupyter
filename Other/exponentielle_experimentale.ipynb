```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fact(n):
    if n>1 :
        return n*fact(n-1)
    else :
        return 1
```

```python
def expo_exp(x,n):
    somme=1
    for i in range(1,n):
        somme=somme+(x**n)/fact(n)
    return somme
```

```python
def erreur_exp(x,n):
    return abs(expo_exp(x,n)-np.exp(x))
```

```python
xp=np.arange(0,1.01,0.01)
n=10
```

```python
plt.plot(xp,expo_exp(xp,n),'r')
plt.plot(xp,erreur_exp(xp,n),'b')
plt.show()
```
