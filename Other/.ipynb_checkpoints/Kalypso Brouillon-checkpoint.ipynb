```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Données 

#nombre d'ateliers
a=1

#nombre de machines
ma=24    

#nombre d'heures de travail à la machine
h=1  

#tonnes de plastique achetés (pour les entreprises Plas1, Plas2 et Plas3)
tP1=1 
tP2=1
tP3=1

#m3 de bois achetés (pour les entreprises Bois1,Bois2 et Bois3 )
mB1=1 
mB2=1
mB3=1
```

# Coûts mensuel de maintenance / production

```python
#Coût mensuel par atelier (mensuel, en euros)
C_Moulage_Atelier = 4000
C_Finition_Atelier = 3000

#Coût mensuel par machine (mensuel)
C_Moulage_Machine = 10000
C_Finition_Machine = 5000

#Coût par heure de travail Machine
C_Moulage_htravail = 50
C_Finition_htravail = 70
```

```python
#fonctions de base
L_Moulage=[]      #liste des fonctions concernant le moulage
L_Finition=[]     #liste des fonctions concernant la finition
L_mensuel=[]      #liste des fonctions de coûts mensuels
L_heure=[]        #liste des fonctions de coûts par heure

#Coût mensuel par atelier
def Moulage_Atelier(n):
    return C_Moulage_Atelier*n*a
L_Moulage.append(Moulage_Atelier)
L_mensuel.append(Moulage_Atelier)

def Finition_Atelier(n):
    return C_Finition_Atelier*n*a
L_Finition.append(Finition_Atelier)
L_mensuel.append(Finition_Atelier)

#Coût mensuel par machine
def Moulage_Machine(n):
    return C_Moulage_Machine*n*ma
L_Moulage.append(Moulage_Machine)
L_mensuel.append(Moulage_Machine)

def Finition_Machine(n):
    return C_Finition_Machine*n*ma
L_Finition.append(Finition_Machine)
L_mensuel.append(Finition_Machine)

#Coût par Heure de Travail Machine
def Moulage_htravail(n):
    return C_Moulage_htravail*n*h
L_Moulage.append(Moulage_htravail)
L_heure.append(Moulage_htravail)

def Finition_htravail(n):
    return C_Finition_htravail*n*h
L_Finition.append(Finition_htravail)
L_heure.append(Finition_htravail)
```

```python
v=np.arange(51)
legendd=' pour '+str(m)+' machines et '+str(a)+' ateliers.'

plt.figure(figsize=(10,10))
for i in L_Moulage:
    plt.plot(v,i(v), label = i.__name__+legendd)
for i in L_Finition:
    plt.plot(v,i(v),label=i.__name__+legendd)

plt.legend()
plt.xlabel('nombre de machine')
plt.ylabel('prix en euros')
plt.title('courbes de tous les coûts mensuels en détail')
```

```python
for i in L_heure:
    plt.plot(v,i(v), label = i.__name__+legendd)
plt.legend()
plt.xlabel('nombre de machine')
plt.ylabel('prix en euros')
plt.title('courbes des coûts par semaine')
```

```python
#fonctions 
L_Totaux=[]


def Total_Atelier(n):
    return Moulage_Atelier(n)+Finition_Atelier(n)
L_Totaux.append([Total_Atelier,a])

def Total_Machine(n):
    return Moulage_Machine(n)+Finition_Machine(n)
L_Totaux.append([Total_Machine,ma])

def Total_htravail(n):
    return Moulage_htravail(n)+Finition_htravail(n)
L_Totaux.append([Total_htravail,h])
```

```python
for i in L_Totaux:
    plt.plot(v,i[0](v),label=i[0].__name__+' pour '+str(i[1])+' '+i[0].__name__[6:])
plt.xlabel('nombre de machine')
plt.ylabel('prix en euros')
plt.legend()
plt.title('courbes des totaux par Atelier , Machine')
```

```python
plt.plot(v,Total_htravail(v),label=Total_htravail.__name__+' pour'+str(i[1])+' '+i[0].__name__[6:])
plt.legend()
plt.xlabel('nombre de machine')
plt.ylabel('prix en euros')
plt.title('courbe de coût total par heure de travail')
```

# Fournisseurs

```python
# Coûts de la tonne de plastique en €
C_Plas1=5100
C_Plas2=5200
C_Plas3=5300

#Coût du m3 de bois en €
C_Bois1=2000
C_Bois2=2150
C_Bois3=2300
```

```python
L_Bois=[]
L_Plas=[]

#écriture des fonctions de base
def Plas1(n):
    return C_Plas1*n*tP1
L_Plas.append([Plas1,tP1])

def Plas2(n):
    return C_Plas2*n*tP2
L_Plas.append([Plas2,tP2])

def Plas3(n):
    return C_Plas3*n*tP3
L_Plas.append([Plas3,tP3])

def Bois1(n):
    return C_Bois1*n*mB1
L_Bois.append([Bois1,mB1])

def Bois2(n):
    return C_Bois2*n*mB2
L_Bois.append([Bois2,mB2])

def Bois3(n):
    return C_Bois3*n*mB3
L_Bois.append([Bois3,mB3])
```

# scénarios

```python
plt.figure(figsize=(10,10))
for i in L_Bois:
    for j in L_Plas:
        plt.plot(v,i[0](v)+j[0](v),label='pour '+str(i[1])+' tonnes de '+i[0].__name__+' et '+str(j[1])+'m3 de '+j[0].__name__)
plt.legend()
plt.xlabel('nombre')
plt.ylabel('coût en euros')
plt.title('scénarios possibles')
```
