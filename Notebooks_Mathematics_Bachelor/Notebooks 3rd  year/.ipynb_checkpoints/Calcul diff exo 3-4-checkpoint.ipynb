```python
import math
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def f3b1(t,C,D):
    return (C*t+D)*np.exp(3*t)-C*np.exp(3*t)

def f3b2(t,C,D):
    return (C*t+D)*np.exp(3*t)
a=20;
b=-20
xp=np.linspace(a,b,100)

def afficherf3b1(C,D):
    return plt.plot(xp,f3b1(xp,C,D),'r')

def afficherf3b2(C,D):
    return plt.plot(xp,f3b2(xp,C,D),'b')
```

```python
afficherf3b1(1,1)
afficherf3b1(2,1)
afficherf3b1(1,2)
afficherf3b1(2,2)
plt.show()
```

```python
afficherf3b2(1,1)
afficherf3b2(1,2)
afficherf3b2(2,1)
afficherf3b2(2,2)
plt.show()
```

```python
def f4a1(t,C,D,E):
    return C*np.exp(2*t)+D*np.exp(3*t)+E*np.exp(6*t)

def f4a2(t,C,D,E):
    return D*np.exp(3*t)-2*E*np.exp(6*t)

def f4a3(t,C,D,E):
    return -1*C*np.exp(2*t)+D*np.exp(3*t)+E*np.exp(6*t)

def afficherf4a1(C,D,E):
    return plt.plot(xp,f4a1(xp,C,D,E),'r')

def afficherf4a2(C,D,E):
    return plt.plot(xp,f4a2(xp,C,D,E),'b')

def afficherf4a3(C,D,E):
    return plt.plot(xp,f4a3(xp,C,D,E),'y')
```

```python
afficherf4a1(1,1,1)
afficherf4a1(2,1,1)
afficherf4a1(1,2,1)
afficherf4a1(1,1,2)
plt.show()
```

```python
afficherf4a2(1,1,1)
afficherf4a2(2,1,1)
afficherf4a2(1,2,1)
afficherf4a2(1,1,2)
plt.show()
```

```python
afficherf4a3(1,1,1)
afficherf4a3(2,1,1)
afficherf4a3(1,2,1)
afficherf4a3(1,1,2)
plt.show()
```

```python
def f4b1(t,C,D,E):      #à corriger
    return C*(t**2/2)*np.exp(2*t)+D*t*np.exp(2*t)+E*np.exp(2*t)

def f4b2(t,C,D,E):
    return C*t*np.exp(2*t)+D*np.exp(2*t)

def f4b3(t,C,D,E):
    return C*np.exp(2*t)

def afficherf4b1(C,D,E):
    return plt.plot(xp,f4b1(xp,C,D,E),'r')

def afficherf4b2(C,D,E):
    return plt.plot(xp,f4a1(xp,C,D,E),'b')

def afficherf4b3(C,D,E):
    return plt.plot(xp,f4a1(xp,C,D,E),'y')
```
