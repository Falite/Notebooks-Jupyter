```python
import math 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def f1(t,C,D):
    return C*(np.cos(3*t)-3*np.sin(3*t))+D*(np.sin(3*t)+3*np.cos(3*t))
def f2(t,C,D):
    return C*2*np.cos(3*t)+D*2*np.sin(3*t)
```

```python
a=-10
b=10
xp=np.linspace(a,b,100)

def afficherf1(C,D):
    return plt.plot(xp,f1(xp,C,D),'r')
    
def afficherf2(C,D):
    return plt.plot(xp,f2(xp,C,D),'b')
```

```python
afficherf1(1,1)
afficherf1(1,2)
afficherf1(2,2)
afficherf1(2,1)
plt.xlabel("abcisses")
plt.ylabel("fonctions")
plt.show()
```

```python
afficherf2(1,1)
afficherf2(1,2)
afficherf2(2,1)
afficherf2(2,2)
plt.xlabel("abcisses")
plt.ylabel("fonctions")
plt.show()
```
