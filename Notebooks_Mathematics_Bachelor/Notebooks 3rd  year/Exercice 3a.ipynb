```python
import math
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def f1 (t,C,D):
    return C*np.exp(-t)+D*np.exp(3*t)
def f2 (t,C,D):
    return 2*C*np.exp(-t)-2*D*np.exp(3*t)
```

```python
a=-10
b=10
xp=np.linspace(a,b,100)

def afficherf1(C,D):
    return plt.plot(xp,f1(xp,C,D),'r')
    
def afficherf2(C,D):
    return plt.plot(xp,f2(xp,C,D),'b')
```

```python
afficherf1(1,1)
afficherf1(1,2)
afficherf1(2,2)
afficherf1(2,1)
plt.show()
```

```python
afficherf2(1,1)
afficherf2(1,2)
afficherf2(2,1)
afficherf2(2,2)
plt.show()
```
