#### multigraph d'une fonction Ã  deux variables

```python
import math 
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
```

```python
def f(x,k):
    return -x+5*np.pi+10*k*np.pi

def c(x):
    return -x
a=-150 ; b=150

xp=np.linspace(a,b,100)

plt.plot(xp,c(xp),'y')
plt.plot(xp,f(xp,1),'r')
plt.plot(xp,f(xp,2),'g')
plt.plot(xp,f(xp,3),'b')
plt.plot(xp,f(xp,4),'p')

plt.show()
```
