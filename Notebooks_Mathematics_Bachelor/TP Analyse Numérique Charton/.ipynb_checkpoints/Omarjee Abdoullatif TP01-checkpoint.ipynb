```python
#préliminaire : définir et afficher la fonction
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
z=-5 ; t=5
def f(x,a):
    return 1/(1+a*x**2)
xp=np.linspace(z,t,100);
plt.plot(xp, f(xp,25),'r');
plt.show()
```

```python
#question1
def Lagrange(x,fx):
    
    n=len(x)-1
    A=np.transpose(np.tile(x,(n+1,1)))
    B=np.tile(range(0,n+1),(n+1,1))
    M=A**B #expliqué juste en dessous
    a=np.linalg.solve(M,fx)
    return a
Lagrange([1,2,3],[0,2,4])
```

```python
#utilité de np.tile, et test du "A**B"
F=np.transpose(np.tile([1,2,3],(3,1)))
print(F) #un exemple de matrice
T=np.tile(range(0,3),(3,1))
print(T) #la matrice des exposants
```

```python
Y=F**T
print(Y) #on retrouve bien la matrice souhaitée : les coefficient sont mis au carré
```

```python
#Question 2
x=np.array([-1,0,1])
from numpy.polynomial import Polynomial
a=-1; b=1;
fx=f(x,3)
P=Polynomial(Lagrange(x,fx))
xp=np.linspace(a,b,100);
plt.plot(xp, f(xp,3),'r');
plt.plot(xp, P(xp), 'g');
plt.show()
```

```python
#Question3 
#on définit une autre fonction qui va prendre le maximum du polynôme P
def maximum(liste):
    maxi = liste[0]
    for i in liste:
        if i >= maxi:
            maxi = i
    return maxi
print(max(abs(f(xp,25)-P(xp)))) #on affiche ensuite l'erreur
```
